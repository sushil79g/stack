```python
yes_count = (file_public['Hobby']=='Yes').sum()
no_count = (file_public['Hobby']=='No').sum()
labels = [ 'Having Hobby', 'Not having any hobby']
sizes = [yes_count, no_count]
explode = (0.1, 0)
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, shadow=True)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
yes_count = (file_public['OpenSource']=='Yes').sum()
no_count = (file_public['OpenSource']=='No').sum()
labels = [ 'having open source contribution', 'no open source contribution']
sizes = [yes_count, no_count]

fig1, ax1 = plt.subplots()
ax1.pie(sizes,labels=labels, shadow=False, autopct='%1.1f%%')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
